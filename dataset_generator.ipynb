```python
# Windowed
dcm_list = [k for k in os.listdir('/content/Train_Data')]
os.chdir('/content/Train_Sinograms_Windowed/')
radon_angle_list = np.arange(0., 181., 1)
for dcm in dcm_list:
    ds = pydicom.read_file('/content/Train_Data/' + dcm)
    img = ds.pixel_array*ds.RescaleSlope + ds.RescaleIntercept # Rescaling PVs from dicom data

    img_windowed = img
    #if type(ds.WindowWidth) != 'DSfloat':
    #  window_width = ds.WindowWidth[0]
    try:
        high_val = ds.WindowCenter + ds.WindowWidth/2 # Windowing image
    except TypeError:
        high_val = ds.WindowCenter[0] + ds.WindowWidth[0]/2 # Windowing image
    try:
        low_val = ds.WindowCenter - ds.WindowWidth/2
    except TypeError:
        low_val = ds.WindowCenter[0] - ds.WindowWidth[0]/2
    img_windowed[np.where(img_windowed>high_val)] = high_val
    img_windowed[np.where(img_windowed<low_val)] = low_val
    
    sinogram = radon(img_windowed, radon_angle_list)
    np.savetxt(dcm[:-4]+'_sino_windowed.csv', sinogram, delimiter=',')
```

```python
import csv
import numpy as np
import os
```

```python
filename = open("stage_2_train.csv", 'r')
file = csv.DictReader(filename)
ID = []
label = []
for col in file:
  ID.append(col['ID'])
  label.append(col['Label'])
```

```python
print(ID[23][:-4])
```

```python
ID_any = []
label_any = []
for k in range(len(ID)):
  if ID[k][-4:]=='_any':
    ID_any.append(ID[k][:-4])
    label_any.append(int(label[k]))
```

```python
ID_any_array = np.array(ID_any)
label_any_array = np.array(label_any)
train_labels = np.column_stack((ID_any_array, label_any_array))
train_labels = train_labels[train_labels[:, 0].argsort()]
```

```python
dcm_list = [k for k in os.listdir('/content/Train_Data')]
ID_subsection = []
labels_subsection = []
for k in range(len(dcm_list)):
  if dcm_list[k][:-4] in train_labels[:, 0]:
    labels_index = np.where(train_labels[:, 0]==dcm_list[k][:-4])
    ID_subsection.append(train_labels[labels_index[0][0]][0])
    labels_subsection.append(train_labels[labels_index[0][0]][1])
```

```python
ID_subsection_array = np.array(ID_subsection)
labels_subsection_array = np.array(labels_subsection)
train_labels_subsection = np.column_stack((ID_subsection_array, labels_subsection_array))
train_labels_subsection = train_labels_subsection[train_labels_subsection[:, 0].argsort()]
```
